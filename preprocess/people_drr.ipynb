```python
# kernel must be msd2
import sys;sys.path.append('../')
import constant as c
import os
import pandas as pd
import warnings; warnings.filterwarnings("ignore")
from util.epsg_transform import *
from shapely.geometry import *
```

```python
# get metro origin/destination df
fpath = "../asset/rawfile/drr/공공자전거 대여소 정보(22.12월 기준)_.csv"
drr = pd.read_csv(fpath) #, encoding = 'cp949')
cols = ['id', 'name', 'gu_nm', 'platplc', 'y', 'x', 'est_yyyymmdd', 'lcd', 'qr']
drr = drr[cols]

columns_to_replace = ['lcd', 'qr']
drr[columns_to_replace] = drr[columns_to_replace].fillna(0)
drr['count'] = drr['lcd'] + drr['qr']

x_5179, y_5179 = coord_tranformation(list(drr.x), list(drr.y), "epsg:4326", "epsg:5179")
drr['x_5179'] = x_5179
drr['y_5179'] = y_5179
drr.to_csv('../asset/preprocess/features/df_drr.csv')
```

```python
drr_2020 = drr[drr['est_yyyymmdd'] < '2020-01-31']
drr_2021 = drr[drr['est_yyyymmdd'] < '2021-01-31']
drr_2022 = drr[drr['est_yyyymmdd'] < '2022-01-31']
drr_2023 = drr[drr['est_yyyymmdd'] < '2023-01-31']
print(len(drr_2020))
print(len(drr_2021))
print(len(drr_2022))
print(len(drr_2023))
```

```python
drr_2020.to_csv('../asset/preprocess/features/df_drr_2020.csv')
drr_2021.to_csv('../asset/preprocess/features/df_drr_2021.csv')
drr_2022.to_csv('../asset/preprocess/features/df_drr_2022.csv')
```
