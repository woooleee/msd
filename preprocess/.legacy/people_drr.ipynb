```python
# kernel must be msd2
import sys;sys.path.append('../')
import constant as c
import os
import pandas as pd
import warnings; warnings.filterwarnings("ignore")
from util.epsg_transform import *
from shapely.geometry import *
```

```python
# get metro origin/destination df
drr_folder_path = os.path.join(c.SSD_PATH,'drr')
drr_fname = '서울시 따릉이대여소 마스터 정보.csv'
drr_path = os.path.join(drr_folder_path, drr_fname)
drr = pd.read_csv(drr_path, encoding = 'euc-kr')
drr = drr.rename(columns = {'주소1': 'rnaddr', '위도': 'y', '경도': 'x'})
drr = drr.drop(columns = ['대여소_ID', '주소2'])
drr = drr[drr['x'] != 0]
x_5179, y_5179 = coord_tranformation(list(drr.x), list(drr.y), "epsg:4326", "epsg:5179")
drr['x_5179'] = x_5179
drr['y_5179'] = y_5179
```
